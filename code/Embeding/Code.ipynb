```python
import os
import os.path
from contract_normalization import Contract_Norm
import numpy as np
from tqdm import tqdm
# words = ['pragma', 'solidity', '^', 'versionliteral', 'contract', '%%%%MultiplicatorX3', 'x3', 'multiplicator', '{', 'address', 'public', '%%%%Owner', 'owner', '=', 'msg', '.', 'sender', 'function', '(', ')', 'public', 'payable', '{', '}', 'function', 'withdraw', '(', ')', 'payable', 'public', '{', 'require', '(', 'msg', '.', 'sender', '==', '%%%%Owner', 'owner', ')', '%%%%Owner', 'owner', '.', 'transfer', '(', 'this', '.', 'balance', ')', '}', 'function', '%%%%Command', 'command', '(', 'address', 'adr', 'bytes', 'data', ')', 'payable', 'public', '{', 'require', '(', 'msg', '.', 'sender', '==', '%%%%Owner', 'owner', ')', 'adr', '.', 'call', '.', 'value', '(', 'msg', '.', 'value', ')', '(', 'data', ')', '}', 'function', 'multiplicate', '(', 'address', 'adr', ')', 'public', 'payable', '{', 'if', '(', 'msg', '.', 'value', '>=', 'this', '.', 'balance', ')', '{', 'adr', '.', 'transfer', '(', 'this', '.', 'balance', '+', 'msg', '.', 'value', ')', '}', '}', '}', '<eof>']  # 替换为您的词汇列表





def get_names():
    files = os.listdir('./ContractCode')
    names = []
    for file in files:
        name = file.replace('.sol', '')
        names.append(name)
    return names


def parser(name):
    '''
        parse the user_input.sol into AST 
    '''
    # cmd = f"java -classpath ./Parse/antlr4.jar:./Parse/target/ \
    #             Tokenize ./ContractCode/{name}.sol ./ParseResult/{name}_"
    cmd = f"java -classpath ./Parse/antlr4.jar:./Parse/target/ \
          Tokenize \"./ContractCode/{name}.sol\" \"./ParseResult/{name}_\""
    # 处理文件名有空格的情况
    os.system(cmd) 
    
def normalizer(name):
    cn = Contract_Norm("./ParseResult", name)    
    return cn.line_span, cn.normalized_tokens



def tokens(name):
    parser(name)
    result = normalizer(name)[1]
    return result

def word_vector(name):
    words = tokens(name)
    word_vectors = {word: model.get_word_vector(word) for word in words}

    # 保存嵌入的向量到文件
    

    # 计算代码嵌入
    code_embedding = sum(word_vectors[word] for word in words if word in word_vectors)
    np.save(f'/home/kaima/LLM/Embed_code715/Embed_result/{name}.npy', code_embedding)
    # 输出结果
    # print("Code Embedding:", code_embedding)


def main():
    names = get_names()
    for name in tqdm(names):
        word_vector(name)

if __name__ == '__main__':
# /home/kaima/LLM/Embed/ContractCode/Onyx Protocol.sol
    main()
    # parser('Earning Farm-2023-08-09')
    # print(normalizer('Earning Farm-2023-08-09')[0])
    # print(normalizer('Earning Farm-2023-08-09')[1][0:10])
    
```

```python
def extract_function_body(tokens, function_name):
    function_body = []
    in_function = False
    brace_count = 0

    for token in tokens:
        if token == function_name and not in_function:
            in_function = True  # 开始提取函数体
            function_body.append(token)
            brace_count = 0  # 重置括号计数
        elif in_function:
            function_body.append(token)
            if token == '{':
                brace_count += 1
            elif token == '}':
                brace_count -= 1
            
            # 如果括号计数回到0，表示函数体结束
            if brace_count == 0 and token == '}':
                break

    return function_body if in_function else None



# 提取函数体
function_name = '%%%%Command'
extracted_body = extract_function_body(word_vectors, function_name)
function_embedding = sum(word_vectors[word] for word in extracted_body if word in extracted_body)

# 输出结果
# print("Extracted Function Body:", extracted_body)
print("Function Embedding:", function_embedding)
```
